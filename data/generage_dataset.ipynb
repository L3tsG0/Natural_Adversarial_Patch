```python

from PIL import Image
from pathlib import Path

from pathlib import Path

# ディレクトリへのパスを定義
images_path = Path('leedsbutterfly/images')
masks_path = Path('leedsbutterfly/segmentations')




```

```python
def extract_by_mask(img_path:list[Path],mask_img_path:list[Path])->None:
    # それぞれのディレクトリ内のファイル名のリストを取得
    images_files = {f.stem: f for f in images_path.iterdir() if f.is_file()}
    masks_files = {f.stem.replace('_seg0', ''): f for f in masks_path.iterdir() if f.is_file()}

    # 両リストの共通のファイル名を基にペアを作成
    paired_files = [(images_files[name], masks_files[name]) for name in images_files if name in masks_files]
    for img_path_,mask_img_path_ in paired_files:
        image = Image.open(img_path_).convert("RGBA")
        mask = Image.open(mask_img_path_).convert("L")
        filename = img_path_.name
        save_filename = Path("leedsbutterfly/images")/filename
        print(save_filename)
        cropped = Image.composite(image, Image.new('RGBA', image.size, (255, 255, 255,0)), mask)

        cropped.save(save_filename)
```

```python
extract_by_mask(images_path,masks_path)
```
